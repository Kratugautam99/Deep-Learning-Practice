```python
import numpy as np
from scipy import signal as sg
from scipy import misc
import matplotlib.pyplot as plt
from PIL import Image
from scipy.ndimage import gaussian_filter
```

```python
I = Image.open('2) Dragon.jpg')
I2 = I.convert('L')
print(f'\n Original type: {I2}')
arr = np.array(I2)
print(f'\n Array type: {arr}')
kernel = np.array([[0,1,0],[-1,4,1],[0,1,0]])
grad = sg.convolve2d(arr,kernel,mode='same',boundary='symm')
grad_biases = np.absolute(grad) + 100
grad_biases[grad_biases>255] = 255
%matplotlib inline
print('GRADIENT MAGNITUDE - Feature Map')
fig,aux = plt.subplots(figsize=(10,10))
aux.imshow(np.absolute(grad),cmap='gray')
plt.show()
```

```python
I = Image.open('2) Dragon.jpg')
I2 = I.convert('L')
arr = np.array(I2)
arr_blurred = gaussian_filter(arr, sigma=1)
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
grad_x = sg.convolve2d(arr_blurred, sobel_x, mode='same', boundary='symm')
grad_y = sg.convolve2d(arr_blurred, sobel_y, mode='same', boundary='symm')
grad = np.sqrt(grad_x**2 + grad_y**2)
plt.figure(figsize=(10,10))
plt.imshow(grad, cmap='gray')
plt.title('Edge Detection using Sobel Operator')
```
